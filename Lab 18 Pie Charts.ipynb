### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
import pandas as pd
import matplotlib.pyplot as plt

# 1.1 Top 5 Databases Respondents Want to Work With
# Splitting and exploding DatabaseWantToWorkWith column
df_exploded = df['DatabaseWantToWorkWith'].dropna().str.split(',').explode()
top_5_databases = df_exploded.value_counts().head(5)

# Plotting the pie chart for top 5 databases
plt.figure(figsize=(8, 8))
top_5_databases.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Top 5 Databases Respondents Want to Work With')
plt.ylabel('')  # Hides the y-label for better visual appearance
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
# 1.2 Top 5 Developer Roles
# Splitting and exploding DevType column
df_devtype_exploded = df['DevType'].dropna().str.split(',').explode()
top_5_devtypes = df_devtype_exploded.value_counts().head(5)

# Plotting the pie chart for top 5 developer roles
plt.figure(figsize=(8, 8))
top_5_devtypes.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Set3.colors)
plt.title('Top 5 Developer Roles')
plt.ylabel('')  # Hides the y-label for better visual appearance
plt.show()
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
import pandas as pd
import matplotlib.pyplot as plt

# 1.3 Operating Systems Used Professionally
# Grouping by 'OpSysProfessional use' and counting occurrences
top_op_sys = df.groupby('OpSysProfessional use').size()

# Plotting the pie chart for operating systems
plt.figure(figsize=(8, 8))
top_op_sys.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.tab10.colors)
plt.title('Operating Systems Used Professionally')
plt.show()
```
